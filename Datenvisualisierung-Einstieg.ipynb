```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Einleseroutine

```python
df = pd.read_csv(### path to csvfile)
df = df.drop( ['docId','filename', 'filename.1', 'article_count', 'article_counts_per_issue', 'article_length_chars', 'article_length_words', 'newspaper', 'headline'], axis=1)
print(df.shape)
df.head()
```

```python
df_year = df.groupby(df.loc[:, 'year']).mean()
df_year.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,6), sharey=True)

plt.style.use('fivethirtyeight')

df_year.iloc[:,21].plot(kind='bar', ax=ax[0], color='orange')
df_year.iloc[:,3].plot(kind='line', ax=ax[1])

ax[0].set(title='Topic Kirche',
          xlabel='Jahr',
          ylabel='Topic Gewichtung')

ax[0].tick_params(axis='x', rotation=45)

ax[0].grid(False, axis='x')

ax[0].annotate(xycoords='axes fraction',
            xy=(0.42, 0.5),
            xytext=(0.5, 0.55),
            s='Hochphase Kulturkampf',
            fontsize=14,
            arrowprops=dict(facecolor='black'))

ax[1].set(title='Topic Eisenbahn',
          xlabel='Jahr',
          ylabel='Topic Gewichtung')

fig.tight_layout()

fig.suptitle('Topic Modeling Säulendiagramm');
#fig.savefig('Topic-Modeling-examples.png')
```

```python
fig, ax = plt.subplots(figsize=(16,6))

plt.style.use('fivethirtyeight')

df_year.iloc[:,21].plot(kind='line', ax=ax, color='orange')
df_year.iloc[:,3].plot(kind='line', ax=ax)

ax.set(title='Topic Kirche / Topic Eisenbahn',
          xlabel='Jahr',
          ylabel='Topic Gewichtung')

ax.tick_params(axis='x', rotation=45)

ax.grid(False, axis='x')

ax.annotate(xycoords='axes fraction',
            xy=(0.42, 0.6),
            xytext=(0.45, 0.65),
            s='Hochphase Kulturkampf',
            fontsize=14,
            arrowprops=dict(facecolor='black'))

fig.tight_layout();

```

```python
df_year_month = df.groupby(df.loc[:, 'year_month']).mean()
df_year_month_cleaned = df_year_month.iloc[:, 6:]
```
